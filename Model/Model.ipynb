```python
# Select features and target for the model
features = df.drop(columns=['RiskScore'], errors='ignore')  # Exclude RiskScore for training
target = df['RiskScore']
```

```python
from sklearn.preprocessing import LabelEncoder

# Apply Label Encoding to categorical features
label_encoders = {}
for col in features.select_dtypes(include=['object', 'category']).columns:
    le = LabelEncoder()
    features[col] = le.fit_transform(features[col])
    label_encoders[col] = le

# Verify that all features are numeric
features.info()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score

```

```python

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train a Random Forest model
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = rf_model.predict(X_test)

# Evaluate model performance
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)


# Display the evaluation metrics
model_performance = pd.DataFrame({
    "Metric": ["Mean Absolute Error (MAE)", "R-Squared (R)"],
    "Value": [mae, r2]
})

```

```python
# Display the evaluation metrics without using tools
print("Random Forest Model Performance:")
print(model_performance)
```

```python
# Error Distribution Analysis
import matplotlib.pyplot as plt

# Calculate residuals (errors) between actual and predicted values
errors = y_test - y_pred

# Plot the error distribution
plt.figure(figsize=(10, 6))
plt.hist(errors, bins=30, edgecolor='black', alpha=0.7)
plt.title("Residuals - Difference between Actual and Predicted RiskScore")
plt.xlabel("Error Value (Actual - Predicted)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Step 1: Define fuzzy membership functions for RiskScore
import skfuzzy as fuzz
import numpy as np

# Define RiskScore fuzzy universe
risk_universe = np.arange(0, 101, 1)

# Define fuzzy membership functions for RiskScore
low_risk = fuzz.trapmf(risk_universe, [0, 0, 33, 50])
medium_risk = fuzz.trimf(risk_universe, [33, 50, 66])
high_risk = fuzz.trapmf(risk_universe, [50, 67, 100, 100])

# Step 2: Map predicted RiskScore values to fuzzy memberships
def calculate_fuzzy_membership(predicted_risks):
    fuzzy_results = []
    for score in predicted_risks:
        memberships = {
            'Low Risk': fuzz.interp_membership(risk_universe, low_risk, score),
            'Medium Risk': fuzz.interp_membership(risk_universe, medium_risk, score),
            'High Risk': fuzz.interp_membership(risk_universe, high_risk, score),
        }
        fuzzy_results.append(memberships)
    return fuzzy_results

# Apply fuzzy membership calculation to predicted RiskScore values
fuzzy_memberships = calculate_fuzzy_membership(y_pred)

# Step 3: Add fuzzy memberships to test data for analysis
test_results = X_test.copy()
test_results['Predicted RiskScore'] = y_pred
test_results['Fuzzy Membership'] = fuzzy_memberships

```

```python
def get_dominant_membership(memberships):
    """Return the risk category with the highest membership value."""
    return max(memberships, key=memberships.get)

# Add the dominant category to the results
test_results['Dominant Risk Category'] = test_results['Fuzzy Membership'].apply(get_dominant_membership)

# Display the updated results with the dominant category
test_results[['Predicted RiskScore', 'Fuzzy Membership', 'Dominant Risk Category']].head(20)
```

```python
# Step: Visualize the fuzzy membership functions and the RiskScore distribution

# Plot the fuzzy membership functions (Low Risk, Medium Risk, High Risk)
plt.figure(figsize=(10, 6))

# Plot Low Risk
plt.plot(risk_universe, low_risk, label="Low Risk", color='blue')
# Plot Medium Risk
plt.plot(risk_universe, medium_risk, label="Medium Risk", color='green')
# Plot High Risk
plt.plot(risk_universe, high_risk, label="High Risk", color='red')

# Title and labels
plt.title('Fuzzy Membership Functions for RiskScore')
plt.xlabel('RiskScore')
plt.ylabel('Membership Value')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
